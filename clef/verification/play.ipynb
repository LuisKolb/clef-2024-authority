```python
import os

from clef.utils.data_loading import AuredDataset


root_path = '../../' # path to github repository root level (where setup.py is located)

data_path = os.path.join(root_path, 'clef2024-checkthat-lab/task5', 'data') # relative to root

filepath = os.path.join(data_path, 'English_dev.json') # relative to root

submission_path = os.path.join(root_path,
                               'clef',
                               'RQ1',
                               'data-out',
                               'setup2',
                               'OPENAI-dev.trec.txt')

config = {
    'preprocess': True,
    'add_author_name': False,
    'add_author_bio': False,
}
ds = AuredDataset(filepath, **config)
ds.add_trec_file_judgements(submission_path, sep='\t')

from clef.verification.models.openai import OpenaiVerifier
from clef.verification.verify import Judge, run_verifier_on_dataset


solomon = Judge(scale=False,ignore_nei=True)
oaiver = OpenaiVerifier()
run_verifier_on_dataset(ds[1:2], oaiver, solomon)
```
